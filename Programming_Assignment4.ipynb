```python
#Question 4
num = input()
sum = 0

for i in num:
  i = int(i)
  sum = sum + (i*i*i)
if int(num) == sum:
    print("Armstrong")
else:
    print("not Armstrong")
```

```python
#Question 5
num = input()
rng = int(input())
inum = int(num)
while inum < rng:
    sum = 0
    for i in str(inum):
        i = int(i)
        sum+= (i*i*i)
    if inum == sum:
        print(inum)
    
    inum+=1
```

```python
#Question 4
def fabi(num):
  if num < 0:
    print("Invalid input")
  lst = [num,num+1]

  for i in range(int(input())):
    nxtterm = lst[-1] + lst[-2]
    lst.append(nxtterm)
  print(lst)


fabi(int(input()))
```
